### Pairplot

El pairplot de Seaborn nos ofrece una rápida mirada a las distribuciones y relaciones entre las variables.

```python
sns.pairplot(df[['surface_covered','surface_uncovered','price_m2','rooms','price','densidad','NBI']])
```

El pairplot es fantástico porque revela inmediatamente algunas características gruesas de la muestra de sumo interés. Por un lado, en la diagonal principal vemos las distribuciones asimétricas y "de colas pesadas" del precio y la superficie. Por otro lado, están las pendientes de las relaciones sugeridas por los gráficos cruzados. Por ejemplo, el precio tiene un relación directa (es decir, positiva) con la superficie (lo que es bastante obvio); pero además, esa relación presenta mayor variabilidad a medida que la superficie aumenta (evidenciado en la forma "cónica" del scatter). 

Otra relación interesante es la que vincula el precio por m2 y la superficie cubierta. A partir de una determinada superficie cubierta (aprox. 75 m2), el precio por m2 de la propiedad tiende a disminuir con su superficie; pero esto no es así para las propiedades menores a ese threshold. Este tipo de comportamiento, tan difícil de captar en modelos lineales, parece perfecto para modelos del tipo Decision Tree y sus derivados.  

Con respecto a las variables que hemos añadido ('densidad', que es la densidad poblacional del radio censal de la propiedad, y 'NBI', que es el porcentaje de hogares con necesidades básicas instatisfechas en dicho radio censal), la que parece ser más relevante es la segunda (que utilizamos aquí como una variable proxy de otra que sería más relevante, que es el ingreso per cápita).

Por último, las variables surface_uncovered y rooms parecen distribuirse en forma relativamente uniforme con respecto a las demás variables, lo que da lugar a pensar en que serán menos relevantes para determinar el precio de una propiedad que las otras.

Lo que estos gráficos no nos presentan es la información en un contexto geográfico. Eso es lo que haremos a continuación.

### Choropleth: precio por m2 según el barrio

A continuación trabajaremos con la información agrupada por barrio, que para eso nos tomamos el trabajo de hacer el feature engineering. Para cada barrio, calculamos precio por m2, superficie y porcentaje de departamentos promedio.

```python
# Elimino columnas que no voy a utilizar
df.drop(columns=['created_on','title','description'],inplace=True)
```

```python
pd.set_option('display.float_format', lambda x: '%.2f' % x)
# Totales de cada barrio
df_mean = pd.DataFrame(df.groupby('barrio')['lat'].count())
df_mean.rename({'lat':'total'},inplace=True,axis=1)
df_mean.reset_index(inplace=True)

# Calculo el precio por m2 promedio de cada barrio
df_mean = pd.merge(df_mean,df.groupby('barrio')['price_m2'].mean(),on='barrio')

# Superficie promedio
barrios_surface = df.groupby('barrio')['surface_total'].mean()
for i in range(len(df_mean)):
    barrio = df_mean.at[i,'barrio']
    df_mean.at[i,'surface_total'] = barrios_surface[barrio]

# Proporcion de departamentos
# Agrupo por barrio y tipo de propiedad (conteo de casos)
property_df = df.groupby(['barrio','property_type'])['lat'].count().unstack().fillna(0)
# Me quedo con los departamentos, calculo proporcion
df_mean = pd.merge(df_mean,property_df,on='barrio')
df_mean['prop_PH'] = df_mean['PH'] / df_mean['total']
df_mean['prop_dept'] = df_mean['Departamento'] / df_mean['total']
df_mean['prop_casas'] = df_mean['Casa'] / df_mean['total']
df_mean.drop(columns=['Departamento','Casa','PH'],inplace=True)
df_mean.head(7)
```

```python
# Importo geopandas, importo mapa de los barrios
import geopandas as gpd
os.chdir('..')
map_df = gpd.read_file('Data/barrios.csv')
os.chdir('Notebooks')
map_df[['barrio','geometry']].head(7)
```

```python
# Creo un geoDataFrame map_price_m2, cruzo con la celda anterior y agrego price_m2 de cada barrio
map_data = map_df[['barrio','geometry']]
map_data = pd.merge(map_data,df_mean,on='barrio')

# A ver...
map_data.head(7)
```

```python
# Creo el Choropleth con Geopandas
fig, ax = plt.subplots(1, figsize=(10, 8))
vmin, vmax = map_data['price_m2'].min(), map_data['price_m2'].max()
map_data.plot(column='price_m2', cmap='Reds', linewidth=0.8, ax=ax, edgecolor='0')

# Un poco de styling
ax.axis('off')
ax.set_title('Precio por m2 segun barrio', fontdict={'fontsize': '18', 'fontweight' : '4'})
sm = plt.cm.ScalarMappable(cmap='Reds', norm=plt.Normalize(vmin=vmin, vmax=vmax))
sm._A = []
cbar = fig.colorbar(sm)
```

### Precio por m2 y distancia a Palermo

El mapa presenta la conocida distribución de precios de propiedades en la Ciudad de Buenos Aires, con su notoria división entre el corredor Norte y corredor Sur. De hecho, esta presentación podría sugerirnos una variable más "amigable con modelos lineales" que explique la relación entre localización geográfica y precio por m2. En efecto, si omitimos Puerto Madero, en el mapa anterior puede notarse que, a medida que nos alejamos de Palermo, la intensidad del color disminuye (es decir, el precio por m2 desciende). 

A continuación investigamos si esta intuición gráfica se ve respaldada por los datos. Para ello, vamos a construir un DataFrame con dos columnas: precio por m2 y distancia a Palermo (específicamente, a Plaza Italia). Luego estudiaremos cuál es la relación entre esas dos variables.

```python
# Coordenadas de Plaza Italia
pi_coord = (-34.581014, -58.421074)
# Definimos una función de distancia euclideana
def distance_pi(lat,lon):
    # Longitud aprox de meridianos / paralelos en metros
    deglen = 110500
    # Distancia euclideana entre dos pares de coordenadas
    delta_x = lat - pi_coord[0]
    delta_y = (lon - pi_coord[1])*np.cos((lat + pi_coord[0]) / 2)
    return deglen*((delta_x)**2 + (delta_y)**2)**(0.5)

```

```python
pd.set_option('display.float_format', lambda x: '%.2f' % x)
df_distPI = df[['lat','lon','price_m2']]
df_distPI['dist_PI'] = distance_pi(df.lat,df.lon)
corr_dist_price = df_distPI[['price_m2','dist_PI']].corr()
print(corr_dist_price)
sns.set(style="white")
fig_dist = sns.lmplot('dist_PI', 'price_m2', data=df_distPI, height=5, aspect=3)
fig_dist2 = sns.scatterplot('dist_PI', 'price_m2', data=df_distPI)
```

Efectivamente, una correlación negativa y una línea de regresión con pendiente negativa describen la relación entre distancia a Plaza Italia y precio por m2.

En el presente apéndice generaremos imágenes y outputs para la realización de mapas interactivos con esta información. Los datos procesados los volcamos en un archivo .js, en un formato listo para ser utilizado en javascript.

### Data para Choropleth

```python
# Lista de variables a las que les voy a crear un mapa
var_list = ['price_m2','surface_total','prop_dept']

# Creo un lista que contiene barrios y valor de la variable
os.chdir('..')
os.chdir('Web')
for var in var_list:
    # Creo lista vacía
    data_list = []
    for i in range(len(map_data)):
        if var != 'prop_dept':
            value = int(map_data.at[i,var])
        else:
            value = round(map_data.at[i,var],4)*100
        list_aux = [map_data.at[i,'barrio'],value]
        data_list.append(list_aux)

    # Le doy formato javascript para exportar
    string_js = 'var '+var+' = '+str(data_list)+'\n\n'+\
            str('for(i=0;i<barrios.features.length;i++) {\n    barrios.features[i].properties.'+var+' = '+var+'[i][1]\n}')+'\n\n'

    # Agrego esta informacion al archivo barrios_data.js
    with open('choroplethBarriosData.js', 'w') as file:
        file.write(string_js)

os.chdir('..')
os.chdir('Notebooks')
```

A continuación creamos los gráficos que aparecerán como popups en nuestros choropleths.

```python
# Creo un histograma para la distribución de precio y superficie en cada barrio
os.chdir('..')
os.chdir('Web/Imgs')

# Creo un histograma para la distribución de precio y superficie en cada barrio
for var in var_list[0:2]:
    os.chdir('Hist '+var)
    for barrio in df.barrio.unique():
        if var == 'price_m2':
            color_hist = '#b30000'
        else:
            color_hist = '#006837'
        plt.hist(df[df.barrio == barrio][var],color=color_hist)
        plt.title(str(var+': '+barrio),{'fontweight':'bold'},pad=10)
        plt.title(str('Muestras: '+str(df[df.barrio == barrio].shape[0])), loc='right')
        plt.xlabel(var)
        plt.ylabel('Propiedades')
        plt.savefig(str(barrio+'.png'))
        plt.close()
    os.chdir('..')

# Creo pie chart para la distribución de propiedades en cada barrio
os.chdir('Pie property type')

for i in range(len(map_data)):
    labels = 'PHs: '+str('%.2f'%(map_data.at[i,'prop_PH']*100))+'%',\
         'Deptos: '+str('%.2f'%(map_data.at[i,'prop_dept']*100))+'%',\
          'Casas: '+str('%.2f'%(map_data.at[i,'prop_casas']*100))+'%'
    sizes = map_data.loc[i,['prop_PH','prop_dept','prop_casas']]
    fig1, ax1 = plt.subplots()
    ax1.pie(sizes, labels=labels,shadow=True, startangle=90)
    ax1.axis('equal')
    plt.title(map_data.loc[i,'barrio'],{'fontweight':'bold'},pad=20)
    plt.savefig(str(map_data.loc[i,'barrio']+'.png'))
    plt.close()

os.chdir('..')
os.chdir('..')
os.chdir('..')
os.chdir('Notebooks')
```
