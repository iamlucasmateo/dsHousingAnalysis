# Estudio de caso: ¿cuánto vale estar cerca del subte?

## Un puzzle estadístico

En el presente notebook analizamos la relacion entre la cercania a una boca de subte y el valor de una propiedad. Es un lugar común la idea de que una propiedad cercana a una boca de un subte tiene, ceteris paribus, un valor mayor que una lejana. En nuestro país, dicho lugar común está inclusive consagrado por ley 23.514 de 1987, que establece que las personas que posean propiedades en un radio de 400 metros de una boca de subte deberán pagar una "contribución de mejoras" (es decir, un impuesto ad hoc) cuando se habilite un tramo nuevo de la red. 
<br><br>
Para analizar esta cuestión vamos a partir de los datos que fueron procesados en el <b>[notebook de Data Processing de Mercado Inmobiliario de CABA](files/Data Preprocessing & Exploration - Mercado Inmobiliario.ipynb)</b>. También haremos uso de la información geográfica disponible en [BA Data](https://data.buenosaires.gob.ar), sobre las bocas de subte en la ciudad de Buenos Aires.

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
import seaborn as sns
```

```python
pd.set_option('display.float_format', lambda x: '%.4f' % x)
os.chdir('..')
df = pd.read_csv('Data/Properati_data_clean.csv',parse_dates=['created_on'],infer_datetime_format=True)
os.chdir('Notebooks')
df.head()
```

```python
# Me quedo solo con las columnas que voy a utilizar

df = df[['lat','lon','barrio','price_m2']]
df.head()
```

```python
# Para cada propiedad, vamos a calcular la distancia a la boca de subte más cercana

# Importo lat-lon de bocas de subte (Fuente: BA Data)
os.chdir('..')
df_subte = pd.read_csv('Data/bocas-de-subte.csv')
os.chdir('Notebooks')
print(f'Total bocas de subte CABA: '+str(df_subte.shape[0]))

# Muestra para chusmear el dataset
df_subte.iloc[0:7,0:5]
```

```python
# Definimos una función de distancia euclideana entre dos puntos
# Esto no tiene en cuenta la curvatura de la tierra, pero para esta escala es irrelevante el error
# Recuérdese que lat y lon son ángulos, por eso se usa el coseno

def distance(lon1,lat1,lon2,lat2):
    # Longitud aprox de meridianos / paralelos en metros
    deglen = 110500
    delta_x = lat1 - lat2
    delta_y = (lon1 - lon2)*np.cos((lat1 + lat2) / 2)
    return deglen*((delta_x)**2 + (delta_y)**2)**(0.5)
```

```python
# Algoritmo para calcular mínima distancia a una boca de subte

# Lo calculamos para cada una de las propiedades en el dataset

for i in range(len(df)):    
    # Coordenadas de la propiedad 
    lon1 = df.loc[i,'lon']
    lat1 = df.loc[i,'lat']
    # Valor inicial min_dist, exageradamente alto
    min_dist = 999999
    
    # Esto de abajo es un control para ir viendo como avanza, cada mil muestras, de ansioso nomas    
#     if abs(i % 1000 - 0) < 0.00001:
#         print(i)
    
    # Calculo distancia a cada boca y me voy quedando con el minimo
    for j in range(0,len(df_subte)):
        lon2 = df_subte.loc[j,'long']
        lat2 = df_subte.loc[j,'lat']
        dist_aux = distance(lon1,lat1,lon2,lat2)
        if dist_aux < min_dist:
            min_dist = dist_aux
            estacion_id = df_subte.loc[j,'id']
    
    # Asigno el minimo hallado y la boca correspondiente a dos columnas nuevas
    df.loc[i,'min_dist'] = min_dist
    df.loc[i,'estacion_id'] = str(estacion_id)
```

```python
# Vemos como nos quedo nuestro df
df.head(7)
```

A continuación vamos a estudiar la relacion entre el precio por m2 y la distancia a una boca de subte. Primero calculamos la correlación entre ambas variables (para toda la muestra) y luego hacemos un scatter con su línea de regresión correspondiente.

```python
corr_dist_subte = df[['price_m2','min_dist']].corr()
print(corr_dist_subte)
sns.set(style="white")
fig_dist = sns.lmplot('min_dist', 'price_m2', data=df, height=5, aspect=3)
fig_dist2 = sns.scatterplot('min_dist', 'price_m2', data=df)
```

Esto nos indica dos cosas: una correlación de Pearson negativa, de -0.19, que no es un valor grande pero sin duda es estadísticamente significativo (dado el gran tamaño de muestra que tenemos). De hecho, podemos calcular el p-value con scipy:

```python
from scipy.stats import pearsonr

results = pearsonr(df.price_m2,df.min_dist)
print(f'Coeficiente de correlación de Pearson: ',results[0])
print(f'p-value: ',results[1])
```

El p-value nos indica la probabilidad de aceptar una hipótesis nula (i.e., ausencia de correlación) que sea falsa. Es bajísimo, es decir, la evidencia indica la existencia de una (leve) correlación negativa entre ambas variables.

Vamos a estudiar el tema más en detalle, barrio por barrio. Los resultados son más que interesantes.

```python
# Creamos un dataframe con los barrios
df_barrios = pd.DataFrame({'barrio':[x for x in df.barrio.unique()]})

# Calculamos correlación en cada barrio y lo agregamos a df_barrios
for i in range(len(df_barrios)):
    barrio = df_barrios.at[i,'barrio']
    # Filtramos el df (original, no el de los barrios) por el barrio en cuestión
    df_filter = df[df.barrio == barrio]
    pearson_coef = pearsonr(df_filter.price_m2,df_filter.min_dist)
    df_barrios.at[i,'corr'] = pearson_coef[0]
    df_barrios.at[i,'p-val'] = pearson_coef[1]
    df_barrios.at[i,'dist_mean'] = df_filter['min_dist'].mean()
    df_barrios.at[i,'obs'] = str(df_filter['lat'].count())
```

A continuación estudiaremos en qué barrios la cercanía al subte implica un mayor valor por m2. Para eso nos enfocamos en el coeficiente de correlación y en el p-value, tomando el habitual valor de corte de 0.05. En primer lugar, veremos para qué barrios la relación no es significativa.

```python
# Barrios con correlación no significativa
df_neutrales = df_barrios[df_barrios['p-val'] > 0.05]
print(f'Hay {len(df_neutrales)} barrios donde el subte no impacta en el precio por m2')
df_neutrales
```

```python
# Barrios con correlación significativa positiva
df_positivos = df_barrios[(df_barrios['p-val'] < 0.05) & (df_barrios['corr'] > 0)]
corr_mean = df_positivos['corr'].mean()
print(f'Hay {len(df_positivos)} barrios con correlación positiva significativa')
print('')
print(f'Correlación positiva promedio: {corr_mean}')
df_positivos
```

```python
# Barrios con correlación significativa negativa
df_negativos = df_barrios[(df_barrios['p-val'] < 0.05) & (df_barrios['corr'] < 0)]
corr_mean = df_negativos['corr'].mean()
print(f'Hay {len(df_negativos)} barrios con correlación negativa significativa')
print('')
print(f'Correlación negativa promedio: {corr_mean}')
df_negativos
```

Los resultados son bastante sorprendentes: para la mayoría de los barrios, la correlación entre el precio del m2 y la cercanía al subte no es significativa (es decir, la cercanía al subte no se encuentra asociada a un precio por m2 mayor). Y en los casos en que esta correlación es significativa, <b>hay casi tantos barrios donde es negativa como barrios donde es positiva</b>. Esto es sorprendente porque significa que hay muchos barrios que contradicen la sabiduría convencional: en ellos, estar cerca del subte está asociado a un <b>menor</b> precio por m2. Cabe resaltar que la correlación positiva promedio es inferior a la negativa y que dentro de los barrios con correlación positiva significativa, hay varios donde el coeficiente es muy cercano a cero (Almagro, San Nicolás, Belgrano).

Más interesante aún es mirar la columna de distancia promedio (dist_mean), que indica la distancia promedio de las propiedades que hay en un barrio al subte (y sirve como proxy de qué tan "atravesado" está un barrio por el subte). En los tres grupos (correlación positiva, neutral y negativa) hay barrios muy atravesados por el subte y barrios muy poco atravesados por el subte. Por ejemplo, entre los barrios con correlación positiva, <b>Chacarita (corr = 0.2712)</b> está completamente atravesado por el subte (línea B) mientras que en <b>Veléz Sarsfield (0.2804)</b> no hay estaciones de subte. Entre los barrios con correlación negativa, <b>Parque Chacabuco (-0.4325)</b> está atravesado por la línea E, mientras que <b>Saavedra (-0.3835)</b> no posee una sola estación de subte. Balvanera es un barrio cercano a casi todas las líneas de subte, mientras que Villa Devoto no tiene ninguna. Sin embargo, en ambos casos la correlación no es estadísticamente significativa.

## Choropleth de correlaciones

Vamos a graficar las conclusiones que hallamos en un mapa, que es otra forma de mostrar la complejidad de la relación entre cercanía al subte y precio por m2.

```python
# Vamos a utilizar geopandas
import geopandas as gpd
os.chdir('..')
# Importo el archivo de BA Data con los polígonos de los barrios
map_df = gpd.read_file('Data/barrios.csv')
os.chdir('Notebooks')
```

```python
# Quiero ver la información que tiene el dataset
map_df.columns
```

```python
# Me quedo con lo que me interesa
map_df = map_df[['barrio','geometry']]
# Mergeo con df_barrios
map_df = pd.merge(map_df,df_barrios,on='barrio')
# A ver...
map_df.head(7)
```

```python
# dist_mean no me interesa
map_df.drop(columns='dist_mean',inplace=True)
```

```python
# Hacemos el mapa, es un choropleth un poco más complicado
# porque necesitamos un color map de dos hues.

from matplotlib import cm
from matplotlib.colors import ListedColormap, LinearSegmentedColormap

# Creo el Choropleth con Geopandas
fig, ax = plt.subplots(1, figsize=(10, 8))
vmin, vmax = map_df['corr'].min(), map_df['corr'].max()

# Creo un color map de calor-frío
bottom = cm.get_cmap('Oranges_r', 128).reversed()
top = cm.get_cmap('Blues', 128).reversed()

newcolors = np.vstack((top(np.linspace(0, 1, 128)),
                       bottom(np.linspace(0, 1, 128))))
newcmp = ListedColormap(newcolors, name='OrangeBlue')

# Ploteo los barrios
map_df.plot(column='corr',cmap=newcmp, linewidth=0.8, ax=ax, edgecolor='0')


# Un poco de styling
ax.axis('off')
ax.set_title('Correlación entre la cercanía al subte y el precio por m2', fontdict={'fontsize': '18', 'fontweight' : '4'})
sm = plt.cm.ScalarMappable(cmap=newcmp, norm=plt.Normalize(vmin=vmin, vmax=vmax))
sm._A = []
cbar = fig.colorbar(sm)
```

Efectivamente, como intuimos viendo las tablas de correlación, el mapa es un verdadero caos: refleja la compleja relación que hay entre cercanía al subte y precio por m2. La intensidad de los colores rojizos indica barrios donde estar lejos del subte se asocia a un mayor precio por m2; mientras que la intensidad de los azules implica que, mientras más cerca del subte, mayor el precio por m2.

## Conclusiones: un puzzle estadístico

En conclusión, la relación entre el precio del m2 y la distancia a una boca del subte está muy lejos de ser unívoca. La fuerza de esta relación puede estudiarse más formalmente a través de tests de significación, pero informalmente podríamos indagar en diferentes razones por las cuales el mito no se cumple. La primera y más obvia es la posible baja representatividad de la muestra. En base a estudios con datasets anteriores, donde observé el mismo fenómeno, creo que eso no es lo que está ocurriendo.

Si aceptamos entonces la representatividad de la muestra, las correlaciones negativas son fáciles de entender; son la sabiduría convencional. Con respecto a las correlaciones positivas, uno podría sentirse tentado de asignarlas al azar o a sesgos muestrales; sin embargo, también podrían estar señalando causalidades más sutiles en la distribución geográfica de los precios. Dos variables positivamente correlacionadas pueden no tener una relación causal directa y sin embargo reflejar una tercera variable que las condiciona a ambas. Es decir, no necesariamente las propiedades más alejadas del subte son más atractivas por el hecho mismo de estar más lejos del subte; más bien, estar lejos del subte podría implicar otras cualidades deseables. A modo de ejemplo -y entrando ya descaradamente en el terreno de la especulación-, en un barrio como Chacarita la cercanía con el subte implica estar lejos de Palermo, que aumenta el atractivo de una propiedad. 

En Villa Urquiza, estar cerca del subte es estar cerca de la zona más céntrica y comercial del barrio, lo que desde el punto de vista residencial no es necesariamente lo más deseable. Sería interesante comparar con los precios de los locales comerciales.

Las dos tendencias podrían estar operando en el caso de los barrios sin correlación; por ejemplo, en un barrio completamente atravesado por el subte como Balvanera, la cercanía a una boca puede implicar cierta comodidad, pero también implica la molestia de vivir en una zona de mucho tránsito permanente. Si estas dos tendencias cotradictorias se compensan, la correlación entre valor y distancia será nula. En Villa Pueyrredón, estar cerca del subte puede ser una ventaja a la hora de ir al centro, pero a la vez implica alejarse de la Gral Paz y sus opciones de transporte.

Naturalmente, el tema es susceptible de un análisis más profundo. Con lo que me siento satisfecho es con la metodología, que podría ser ampliada a otras variables mapeadas por el Gobierno de la Ciudad (espacios verdes, sitios posibles de anegamiento, comisarías). Estos cruces de información podrían dar lugar a mejores estimaciones de los precios de los inmuebles en la Ciudad de Buenos Aires.

### Exportando la información para mapas interactivos

A continuación vamos a exportar la información aquí procesada para construir mapas web interactivos con ella. 

```python
# Vamos a trabajar con nuestro map_df que usamos más arriba
# p-val no me interesa, solo si es significativo o no
map_df['signif'] = np.where(map_df['p-val'] < 0.05,True,False)
# Dropeo p-val
map_df.drop(columns='p-val',inplace=True)
```

```python
# Vamos a asignarle un color a cada barrio para graficar

# Definimos nuestros colores y nuestros límites
color_pos_list = ['fff7ec','#fee8c8','#fdd49e','#fdbb84','#fc8d59','#ef6548','#d7301f','#b30000'] # rojos
color_neg_list = ['#f7fbff','#deebf7','#c6dbef','#9ecae1','#6baed6','#4292c6','#2171b5','#08519c'] # azules
lim_pos_list = [0,0.1,0.15,0.2,0.25,0.3,0.35,0.5]
lim_neg_list = [0,-0.1,-0.2,-0.25,-0.3,-0.35,-0.45,-0.6]

# La siguiente función asigna color según la correlación

def color_corr(signif,corr):
    if signif == False:
        return '#ffffff' # si no es significativa, color blanco
    else:
        if corr > 0:
            index = 0
            for i in range(len(lim_pos_list)):
                if corr > lim_pos_list[index]:
                    index += 1
                else: 
                    break
            return color_pos_list[index]
        else:
            index = 0
            for i in range(len(lim_neg_list)):
                if corr > lim_neg_list[index]:
                    index += 1
                else: 
                    break
            return color_neg_list[index]
```

```python
# Agregamos colores al dataframe para cada barrio
for i in range(len(map_df)):
    map_df.at[i,'color'] = color_corr(map_df.at[i,'signif'],map_df.at[i,'corr'])

# A ver...
map_df.head(7)
```

```python
# Por último exportamos muestras individuales cercanas a la linea B,
# con las que haremos un examen detallado.

df_lineaB = df[df.barrio.isin(['VILLA URQUIZA','VILLA ORTUZAR','CHACARITA','VILLA CRESPO','PARQUE CHAS','PATERNAL'])]
df_lineaB = df_lineaB[['price_m2','lat','lon']]
df_lineaB.reset_index(inplace=True,drop=True)
```

```python
list_price = [0,1500,2000,2500,3000,3500,4000,5000]
list_grayscale = ['#f0f0f0','#d9d9d9','#bdbdbd','#969696','#737373','#525252','#252525','#000000']

def price_grayscale(price):
    index = 0
    for i in range(len(list_price)-1):
        if price > list_price[i]:
            index += 1
        else:
            break
    return list_grayscale[index]  
```

```python
lineaB_list = []

for i in range(len(df_lineaB)):
    price = df_lineaB.at[i,'price_m2']
    color = price_grayscale(price)
    lineaB_list.append(['%.2f'%price,df_lineaB.at[i,'lat'],df_lineaB.at[i,'lon'],color])
    
with open('subteLineaB.txt', 'w') as file:
        file.write(str(lineaB_list))
```
